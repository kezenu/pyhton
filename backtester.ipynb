```python
import pandas as pd
import mplfinance as mpf

# === Load Data ===
df = pd.read_csv('EURUSD_M5.csv')
df['time'] = pd.to_datetime(df['time'])
df.set_index('time', inplace=True)

# === Replay Settings ===
window_size = 30  # jumlah candle yang ditampilkan
start = window_size

# === Replay Loop ===
while start < len(df):
    data_window = df.iloc[start-window_size:start]
    
    mpf.plot(data_window, type='candle', style='charles', title=f'Candle ke-{start}',
             ylabel='Harga', volume=False)
    
    cmd = input("Tekan ENTER untuk lanjut, atau ketik 'q' untuk keluar: ")
    if cmd.lower() == 'q':
        break
    
    start += 1
```
